# T-Test
**Definition:** T-Test is used to compare means when the population variance is unknown or the sample size is small.
---
## Independent T-Test
**Example:** Compare **Glucose ** of **positive vs Negative** person.
**Hypotheses: **
- HO: Glucose distribution of positive = Glucose distribution of negative
- H1: Glucose distribution of positive != Glucose distribution of negative

```python
# T-Test 
from scipy import stats
import statsmodels.api as sm
p_pos = df[df['Outcome'] == 1]['Pregnancies']
p_negative = df[df['Outcome'] == 0]['Pregnancies']
```

```python
p_negative
```

```python
t_stat,p_val = stats.ttest_ind(p_pos,p_negative)
print("Independent T-test statistic:",t_stat)
print("P-value:",p_val)
if p_val < 0.05:
    print("Reject null hypothesis:")
else:
    print("Accept null hypothesis:")
```
